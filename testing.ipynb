```python
# funcion que devuelve los indices donde un array es distinto a otro
#recorre los dos arrays y compara si son iguales, si no lo son guarda el indice
def diff(first, second):
        listaa=[]
        for i in range(len(second)):

            if first[i] != second[i]:
                listaa.append(i)

                break
        return listaa
```
