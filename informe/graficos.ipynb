```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
#descargar datos
path='../results/OneHot/metadata_train.npy'
metadata_train = np.load(path, allow_pickle=True)
path='../results/OneHot/metadata_test.npy'
metadata_test = np.load(path, allow_pickle=True)

#unir datos
metadata = np.concatenate((metadata_train, metadata_test), axis=0)

#X es el vector de caracteristicas que son todas las columnas menos la ultima
X = metadata[:, :-1]
#y es la ultima columna
y = metadata[:, -1]

#Hacer PCA de X y graficar
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
#Y es el precio
# hacer un scatter plot de X_pca y poner Y como color de los puntos siendo el color mas rojo el precio mas alto
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm')
plt.colorbar()
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
#label de colorbar que diga precio




```

```python
#elegir un subset de datos de metadata y graficar (aleatorio )
subset= metadata[np.random.choice(metadata.shape[0], 1000, replace=False), :]
X = subset[:, :-1]
y = subset[:, -1]
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
#hacer figura de 10,10
plt.figure(figsize=(10, 10))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm')
plt.colorbar()
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```
